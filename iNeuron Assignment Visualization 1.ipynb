# Assignment : Visualization 1

In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
# read the dataset from sklearn datasets
data = datasets.load_iris()
```

```python
# separate the data to decompose and target
df_data = data.data
df_target = data.target
```

```python
# convert the dataset into 3 dimension
pca = decomposition.PCA(n_components=3)
pca.fit(df_data)
X = pca.transform(df_data)
X
```

```python
# construct the dataframe
df_ = pd.DataFrame(X,columns=["PC1","PC2","PC3"])
df_['Class'] = df_target
df_.head()
```

```python
# replace the number with text class
di = {0:'setosa',
      1:'versicolor',
      2:'virginica'}
df_ = df_.replace({"Class": di})
df_.head()
# plot the graph 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
lst_color = ['r','b','y']
di = ['setosa','versicolor','virginica']
for color,class_ in zip(lst_color,di):
    df__ = df_[df_['Class'] == class_]
    ax.scatter(df__.PC1, df__.PC2, df__.PC3, marker='o',color = color)
ax.legend(di)
plt.title("Iris data 3D with Class")
plt.show()
```
